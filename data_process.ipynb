```python
import pandas as pd
```

```python
columns_names = ["user_id", "item_id", "rating", "timestamp"]
ratings = pd.read_csv("ml-100k/u.data", sep="\t", header=None, names=columns_names)
ratings
```

```python
ratings.describe()
```

```python
columns_names = ["user_id", "age", "gender", "occupation", "zip_code"]
users = pd.read_csv("ml-100k/u.user", sep="|", header=None, names=columns_names)
users
```

```python
columns_names = ["item_id", "title", "release_date", "video_release_date","imdb_url", 
                "unknown", "action", "adventure", "animation", "children", "comedy", "crime", "documentary", "drama", 
                "fantasy", "filmnoir", "horror", "musical", "mystery", "romance", "scifi", "thriller", "war", "western"]
movies = pd.read_csv("ml-100k/u.item", sep="|", header=None, names=columns_names, encoding="ISO-8859-1")
movies
```

```python
df1 = pd.merge(ratings, users, on="user_id")
df1
```

```python
df2 = pd.merge(df1, movies, on="item_id")
df2
```

```python
df2.to_csv("recommendation_full.csv")
```

```python

def set_type(row):
    if row["action"] == 1:
        return "action"
    if row["adventure"] == 1:
        return "adventure"
    if row["animation"] == 1:
        return "animation"
    if row["children"] == 1:
        return "children"
    if row["comedy"] == 1:
        return "comedy"
    if row["crime"] == 1:
        return "crime"
    if row["documentary"] == 1:
        return "documentary"
    if row["drama"] == 1:
        return "drama"
    if row["fantasy"] == 1:
        return "fantasy"
    if row["horror"] == 1:
        return "horror"
    if row["musical"] == 1:
        return "musical"
    if row["mystery"] == 1:
        return "mystery"
    if row["romance"] == 1:
        return "romance"
    if row["scifi"] == 1:
        return "scifi"
    if row["thriller"] == 1:
        return "thriller"
    if row["war"] == 1:
        return "war"
    if row["western"] == 1:
        return "western"
    return "unknown"
```

```python
df2["type"] = df2.apply(set_type, axis=1)
```

```python
df2.head()
```

```python
df3 = df2[["user_id", "item_id", "rating", "age", "gender", "occupation", "zip_code", "title", "release_date", "type"]]
df3.head()
```

```python
df3 = df3.rename_axis('rid')
df3.to_json("recommendations.json", orient="records")
```
