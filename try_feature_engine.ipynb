```python
import pandas as pd

ratings = pd.read_csv("new_cleaned_rating_data.csv")

ratings = ratings[["UserID", "MovieID", "minute", "Rating"]]
ratings.head()
```

```python
ratings["minute"] = ratings["minute"].apply(lambda x: int(x[:-4]))
ratings
```

```python
ratings.describe()
```

```python
from feature_engine.imputation import MeanMedianImputer

mean_imputer = MeanMedianImputer("mean")
mean_imputer.fit(ratings)
mean_filled_ratings = mean_imputer.transform(ratings)
mean_filled_ratings
```

```python
mean_filled_ratings.describe()
```

```python
median_imputer = MeanMedianImputer("median")
median_imputer.fit(ratings)
median_filled_ratings = median_imputer.transform(ratings)
median_filled_ratings
```

```python
median_filled_ratings.describe()
```

```python
from feature_engine.selection import DropHighPSIFeatures

psi = DropHighPSIFeatures()
psi.fit_transform(median_filled_ratings)
```

```python
from feature_engine.selection import DropCorrelatedFeatures
dcf = DropCorrelatedFeatures(threshold=0.7)
dcf.fit_transform(mean_filled_ratings)
```

```python
import pandas as pd
from feature_engine.encoding import OneHotEncoder

recommendations = pd.read_csv("recommendations.csv")
ohe = OneHotEncoder()
ohe.fit(recommendations[["type"]])
feat = ohe.transform(recommendations[["type"]])
feat
```

```python
from feature_engine.datetime import DatetimeFeatures
from feature_engine.imputation import DropMissingData
dmd = DropMissingData(variables=["release_date"])
recommendations_ = dmd.fit_transform(recommendations)
dtf = DatetimeFeatures(features_to_extract=["year"])
recommendations_["release_year"] = dtf.fit_transform(recommendations_[["release_date"]])
recommendations_
```
